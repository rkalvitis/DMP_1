#3. Datu sagatavošana

```python
# Ielasa datus pandas dataframe objektos
# training_data tiks izmantoti neironu tīkla apmācīšanai
training_data = pd.read_csv("Iris_train.csv")
# validation_data tiks izmantoti neironu tīkla precizitātes pārbaudīšanai
validation_data = pd.read_csv("Iris_validate.csv")
#validation_data = pd.read_csv("Iris_modified.csv") # 150 entries (pilna Iris datu kopa (https://github.com/siddharthjain1611/Iris_dataset/blob/master/Iris.csv))
#validation_data = pd.read_csv("Iris_different.csv") # 23 entires (Iris_modified - Iris_train)


# test_data izmantos pasniedzējs rezultātu pārbaudei, atstāt aizkomentētu
# test_data = pd.read_csv("Iris_test.csv")

# Pārveido datus neironu tīklam piemērotā formā
X_train = training_data.drop('Species', axis=1)
y_train = training_data['Species']
# Teksta etiķešu kodēšana uz skaitļiem
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)

# Tas pats darīts priekš validācijas datu kopas
X_validate = validation_data.drop('Species', axis=1)
y_validate = validation_data['Species']
y_validate_encoded = encoder.transform(y_validate)
```

#4. Neironu tīkla definēšana

```python
# Definē neironu tīklu

#model = Sequential()
#model.add(Dense(10, input_dim=4, activation='relu')) # Slānis ar 10 neironiem
#model.add(Dense(8, activation='relu')) # Vēl viens slānis
#model.add(Dense(3, activation='softmax')) # Izvades slānis (3 klases)



model = Sequential()
model.add(Dense(12, input_dim=4, activation='relu')) # Palielinām neironu skaitu
model.add(Dropout(0.2))  # Pievienojam Dropout slāni
model.add(Dense(8, activation='relu')) # Saglabājam šo slāni
model.add(Dropout(0.2))  # Pievienojam vēl vienu Dropout slāni
model.add(Dense(3, activation='softmax')) # Izvades slānis paliek nemainīgs
```

#5. Neironu tīkla apmācīšana

```python
# Apmāca neironu tīklu klasificēt īrisus pēc to ziedu raksturlielumiem

#model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
#model.fit(X_train, y_train_encoded, epochs=150, batch_size=10)

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train_encoded, epochs=150, batch_size=5)
```

# 6. Neironu tīkla pārbaudīšana.
Jāizdrukā visu pārbaužu rezultāti viegli saprotamā formā (12 rindiņas, pa vienai katram validācijas kopas piemēram), kā arī kopējais precizitātes novērtējums.  
 Noformējuma piemērs:  
 1. rindiņa: klasifikatora rezultāts: Iris-setosa; reālā suga: Iris-setosa, pareizi  
 ...
 12. rindiņa: klasifikatora rezultāts: Iris-versicolor; reālā suga: Iris-virginica, nepareizi  
 Precizitāte: 58% (7/12)

```python
# Veic prognozes validācijas datiem
predictions = model.predict(X_validate)
predicted_classes = np.argmax(predictions, axis=1)
true_classes = y_validate_encoded

# Izveido atbilstošus nosaukumus klasēm
predicted_class_labels = encoder.inverse_transform(predicted_classes)

# Izdrukā katras rindiņas rezultātus
count = 0
for i in range(len(predicted_class_labels)):
    count = count +1
    correctness = "pareizi" if predicted_class_labels[i] == y_validate.iloc[i] else "nepareizi"
    print(f"{count}. rindiņa: klasifikatora rezultāts: {predicted_class_labels[i]}; reālā suga: {y_validate.iloc[i]}, {correctness}")

# Aprēķina un izdrukā kopējo precizitāti
correct_predictions = np.sum(predicted_class_labels == y_validate.values)
total_predictions = len(predicted_class_labels)
accuracy = correct_predictions / total_predictions
print(f"Precizitāte: {accuracy * 100:.2f}% ({correct_predictions}/{total_predictions})")
```
